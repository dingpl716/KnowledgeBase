```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
folder = '/Users/peiling/Projects/dingpl716/ex_pid/sim_output'
def get_data_frames(file_name) -> any:
    file_path = f"{folder}/{file_name}"
    df = pd.read_csv(file_path, parse_dates=['datetime'])
    df.set_index('datetime', inplace=True)
    return df
```

```python
values = [0.5, 1, 2]
df_map = {}
for kp in values:
    for ki in values:
        for kd in values:
            for kf in values:
                file_name = f"k_{kp}_i_{ki}_d_{kd}_f_{kf}.csv"
                df_map[file_name] = get_data_frames(file_name)                
```

```python
#define number of rows and columns for subplots
nrow=27
ncol=3

# make a list of all dataframes 
fig, axes = plt.subplots(nrow, ncol, figsize=(24,135))
plt.subplots_adjust(hspace=0.5)

# plot counter
count = 0
names = list(df_map.keys())
for r in range(nrow):
    for c in range(ncol):
        name = names[count]
        df_map[name]["current_velocity"].plot(title=name, ax=axes[r,c])
        count+=1
```
